# Selection Demo
Show that selections in a results table are shown in Aladin and vice versa.  

As in Missions MAST, selections in Aladin can be made by clicking on the footprint edges or by selecting a region. Note that with a recent change in Aladin, region selections grab any footprints that have at least one vertex in the region.
## Imports
Imports include `mast_search_helpers` which is a local module to help switching among backend environments and loaded static result sets.

```python
import astropy.units as u
from astropy.coordinates import SkyCoord

from astroquery.mast import MastMissions
from astropy.table import Table
from mast_aladin import MastTable, AppSidecar, MastAladin

from mast_search_helpers import saved_roman_test_results
```

## "Search" for Roman data
Load a previously-saved result set from the Roman test environment.

```python
target = SkyCoord(ra=270, dec=65.8, unit='deg')
roman_obs = saved_roman_test_results()
```

## Create Aladin and the Table view
- Create a `MastAladin` instance and display it off to the right.
- Create a `MastTable` instance to provide a tabular view of search results.
- Use `add_table` Load the result into Aladin to see the footprints.

The `add_table` arguments can likely be simplified when we refactor `MastTable`.  The important piece of information to pass to Aladin is the `item_key` which is the name of the column that contains a unique identifier for the results.  In this example, the `MastTable` has made some educated guesses about the unique id column based on its knowledge of Missions MAST results.  Without knowing a unique identifier, we would not be able to communicate between widgets about selections.

```python
aladin = MastAladin(fov=1.0)
apps_initialized = AppSidecar.open(aladin, anchor='split-right')
AppSidecar.resize_all(800)

mast_table_roman_obs = MastTable(roman_obs, app=aladin)

roman_obs_layer = aladin.add_table(roman_obs, name='Roman Observations', color='#FF6600', 
                             ra_field="ra_ref", dec_field="dec_ref",
                            mast_item_key=mast_table_roman_obs.item_key)
aladin.target = f'{target.ra.deg} {target.dec.deg}'

mast_table_roman_obs
```

```python
aladin.selection('rectangle')
```
