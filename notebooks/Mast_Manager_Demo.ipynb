## Mast Manager Demo
The goal of this notebook is to demonstrate how to utilize the MastManager to its full potential. It is the central hub that facilitates all communication between apps and services that we support. 

```python
from mast_aladin import MastManager, MastAladin
from jdaviz import Imviz
```

### Initialize Mast Manager
Create the MastManager instance. This should be done early in the notebook. Re-running this cell will wipe any registered apps. Do so at your own peril

```python
mm = MastManager()
```

### Initialize Apps
Create the applications you want to use nad register them with Mast Manager. You can also create new applications later and Mast Manager will update accordingly once they are registered

```python
aladin = MastAladin()
mm.register_app(aladin, "aladin")
```

```python
viz = Imviz()
mm.register_app(viz, "viz")
```

You can retrieve the list of registered apps at any point by querying the Mast Manager API 

```python
mm.apps
```

You can retrieve the app instances as well by querying the Mast Manager API

```python
viz = mm.apps["viz"]
```

### Display and App Layout
Mast Manager makes use of the Sidecar library to enable interactive display layouts of the registered apps and supported plugins. You can either use the default layout or define a custom one

```python
mm.display()
```

```python
mm.close()
```

If you decide to use a custom layout, ensure that all registered apps you want to load are referenced. You can also include a custom layout for the plugins UI

**NOTE:** The extact interface for custom layouts is still in development.

```python
mm.display({
    "viz": "split-right",
    "aladin": "split-bottom",
    "plugins": "split-bottom"
})
```

If you like the layout you have created and want to save it load again, you can do so via. layout templates

```python
mm.save_layout("my_favorite_layout")
```

```python
mm.load_layout("my_favorite_layout")
```

### Configure Plugins 
By default the Mast Manager enables a list of default supported plugins. You can retrieve the list of enabled plugins and enable or disable plugins through the UI

```python
mm.plugins
```

```python
mm.disable_plugin("viewer_sync")
```

```python
mm.enable_plugin("viewer_sync")
```

You can also use this API to call the plugins directly, though it is suggested that you use the UI if possible

```python
viewer_sync_plugin = mm.plugins["viewer_sync"]
viewer_sync_plugin.sync("aladin", "viz")
```
